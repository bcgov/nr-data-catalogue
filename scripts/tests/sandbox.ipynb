```python
# os.chdir('..')
print(f"Current working directory: {os.getcwd()}")

from config import *

# Endpoint URL that works in Postman
url = "https://nr-data-catalogue-dev.apps.emerald.devops.gov.bc.ca/api/v1/tables?limit=99999&fields=columns,tags"


# API request headers
headers_get = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
}

# Make the GET request to retrieve table data
response = requests.get(url, headers=headers_get)

# Check the response status
if response.status_code == 200:
    data = response.json()  # Get the JSON response
    tables_data = []
    
    # Extract tables and columns with tags
    for table in data.get('data', []):  # Ensure 'data' exists
        table_name = table.get('fullyQualifiedName')
        columns = table.get('columns', [])
        
        # Loop through each column and get tags
        for column in columns:
            column_name = column.get('name')
            column_tags = [tag['tagFQN'] for tag in column.get('tags', []) if 'tagFQN' in tag]  # Ensure tagFQN exists
            
            # Append table name, column name, and tags to the list
            tables_data.append([table_name, column_name, ', '.join(column_tags)])

    # Convert the list into a pandas DataFrame
    df = pd.DataFrame(tables_data, columns=["Table Name", "Column Name", "Column Tags"])
    print(df)

    # Save the DataFrame to a CSV for further use
    df.to_csv("table_column_tags.csv", index=False)

else:
    print(f"Failed with status code {response.status_code}")
    print(response.text)
```
